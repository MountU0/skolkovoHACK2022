```python
import re

'''
На вход подается строка
Выход - None, если паттерн не найден, иначе список, в котором находятся подходящие гражданства
'''
def get_citizenship(str):
    pattern = r'Гражданство[:]? (РФ|СНГ)?[,]?[ ]?(РФ|СНГ)?'
    RF_pattern = r'РФ'
    CIS_pattern = r'СНГ'
    match = re.search(pattern, str, re.IGNORECASE)
    answer = []
    if match is not None:

        new_str = match.group(0)

        RF_match = re.search(RF_pattern, new_str, re.IGNORECASE)
        CIS_match = re.search(CIS_pattern, new_str, re.IGNORECASE)

        if RF_match is None and CIS_match is None:
            return answer.append(2)
        
        if RF_match is not None:
            answer.append(0)
        
        if CIS_match is not None:
            answer.append(1)
        
        return  answer

    return None
    
    

'''
На вход подается строка
Выход - Стаж вождения [int]
        функция, которая вернет True, если передаваемый параметр будет не меньше, чем год в строке
'''
def get_driving_experience(str):
    pattern = r'Стаж вождения[ ]?[:]?[ ]?( от)?(не менее)? \d{1,2}'


    year_pattern = r'\d{1,2}'

    match = re.search(pattern, str, re.IGNORECASE)


    if match is not None:

        new_str = match.group(0)


        year_match = re.search(year_pattern, new_str, re.IGNORECASE)

        if year_match is None:
            return None, lambda x: None
        
        year = int(year_match.group(0))

        return year, lambda x: x >= year

'''
На вход подается строка
Выход - Пол: Мужской\Женский
'''

def get_sex(str):

    pattern_man = r'(мужчина)[.| |,|;|"]|(муж)[.| |,|;|"]'

    pattern_woman = r'(женщина)[.| |,|;|"]|(жен)[.| |,|;|"]'

    matc_man = re.search(pattern_man, str, re.IGNORECASE)
    #Мужчина
    if matc_man is not None:
        return 2
    
    matc_woman = re.search(pattern_woman, str, re.IGNORECASE)

    if matc_woman is not None:
        return 1
    
    return 0
'''
На вход подается строка
Выход - список, в котором возраст (или диапозон возрастов ) [Все приведено в int]
        функция, проверяющая, что возраст подходит 
'''
def get_age(str):
    pattern = r'(возраст| возрасте)[ ]?[:]?[-]?(до)?(от)?[ ]?[\d]*[ ]?[-]?[ ]?[\d]*'
    pattern_range = r'(\d{1,2})[ ]?-[ ]?(\d{1,2})'
    pattern_range_no_dash = r'(\d{1,2} \d{1,2})'
    pattern_inequality = r'(до|от)'
    pattern_year = r'\d{1,2}'

    match = re.search(pattern, str, re.IGNORECASE)
    
    if match is not None:
        new_str = match.group(0)
        range_match = re.search(pattern_range, new_str, re.IGNORECASE)
        if range_match is not None:
            range_list = list(map(int, range_match.group(0).replace(' ', '').split('-')))
            return range_list, lambda x: range_list[0] <= x <= range_list[1]
        

        range_no_dash_match = re.search(pattern_range_no_dash, new_str, re.IGNORECASE)
        if range_no_dash_match is not None:
            range_no_dash = list(map(int, range_no_dash_match.group(0).split(' ')))
            return range_no_dash, lambda x: range_no_dash[0] <= x <= range_no_dash[1]

        inequality_match = re.search(pattern_inequality, new_str, re.IGNORECASE)
        year = int(re.search(pattern_year, new_str, re.IGNORECASE).group(0))
        if inequality_match is not None:
            inequality_str = inequality_match.group(0)

            if inequality_str == 'до':
                return [year], lambda x: x < year
            else:
                return [year], lambda x: x >= year
        else:
            return[year], lambda x: x >= year

    return None, lambda x: 1
    

def get_license(str):
    pattern = r'(категории|категория) [A]?[,]?[ ]?[B]?[,]?[ ]?[C]?[,]?[ ]?[D]?[,]?[ ]?[BE]?[,]?[ ]?[CE]?[,]?[ ]?[DE]?[,]?[ ]?[TM]?[,]?[ ]?[M]?[,]?[ ]?[TB]?( |;|<|,)'

    driverLicense_dict = dict()
    match = re.search(pattern, str, re.IGNORECASE)
    lst = []
    if match is not None:
        new_str = match.group(0).replace(',', ' ').replace('<', ' ').replace('>', ' ').replace('.', ' ').replace('/', ' ')
        if re.search('(a)',str.lower()): lst.append('A')
        if re.search('(b)',str.lower()): lst.append('B')
        if re.search('(c)',str.lower()): lst.append('C')
        if re.search('(d)',str.lower()): lst.append('D')
        if re.search('(e)',str.lower()): lst.append('E')
        if re.search('(be)',str.lower()): lst.append('BE')
        if re.search('(ce)',str.lower()): lst.append('CE')
        if re.search('(de)',str.lower()): lst.append('DE')
    
        #data_candidates['DriverLicense'] = data_candidates['DriverLicense'].apply(lambda x: driverLicense_dict[x])

    return lst
```

```python
sex_dis = []
age_dis = []
citizenship_dis = []
license_dis = []
#print(get_citizenship(str))
for index, row in data_candidates.iterrows():
    sex = get_sex(row['Description'])
    if sex != 0:
        if row['Sex'] == 0:
            sex_dis.append(0)
        else:
            sex_dis.append(int(sex == row['Sex']))

    else:
        sex_dis.append(1)
    #print(index)
    age, func_age = get_age(row['Description'])
    age_dis.append(func_age(row['Age']))

    #сравнение с показателем у кандидата

    # citizenship_dis.append(get_citizenship(row['Description']))
    citizenship = get_citizenship(row['Description'])

    if citizenship is None:
        citizenship_dis.append(1)

    elif row['Citizenship'] in citizenship:
        citizenship_dis.append(1)
    else:
        citizenship_dis.append(0)


    description_license = get_license(row['Description'])
    #Проверка на вхождение лицензий
    equality_licenses = 1
    for i in description_license:
        if i not in row['DriverLicense']:
            equality_licenses = 0
            break
    license_dis.append(equality_licenses)

data_candidates['sex_0_1'] = sex_dis
data_candidates['age_0_1'] = age_dis
data_candidates['license_dis'] = license_dis
data_candidates['citizenship_dis'] = citizenship_dis
```
